### Train-Test Split

```python
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train = df.drop(['Attrition'],axis = 1)
test = df['Attrition']
```

```python
train = pd.get_dummies(train, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=0.25,shuffle=True,stratify = test, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Model Building

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
```

### Random-Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF = RandomForestClassifier()
```

```python
RF.fit(X_train,y_train)
```

```python
RF.score(X_test,y_test)
```

```python
scores = cross_val_score(RF, X_train, y_train, scoring ="roc_auc", cv = 5)
roc_auc_RF = np.mean(scores)
roc_auc_RF
```

```python
y_pred = RF.predict(X_test)
```

```python
roc=roc_auc_RF
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

model_results = pd.DataFrame([['Random Forest', acc,prec,rec, f1,roc]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score','ROC'])

model_results
```
